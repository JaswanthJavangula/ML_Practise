```python
# Install dependencies as needed:

import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "Loan.csv" # Corrected file path

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS, # Adapter should be the first argument for load_dataset
  "lorenzozoppelletto/financial-risk-for-loan-approval",
  file_path
)
df = df.head(5500)
print("First 5 records:", df.head())
```

```python
import pandas as pd
import numpy as np
```

```python
df.tail(3)
```

```python
df.columns
```

```python
df["RiskScore"]
```

```python
X = df[[
    "Age",
    "AnnualIncome",
    "CreditScore",
    "LoanAmount",
    "LoanDuration",
    "MonthlyDebtPayments",
    "DebtToIncomeRatio",
    "TotalDebtToIncomeRatio",
    "NetWorth",
    "RiskScore"
]]
y= df["LoanApproved"]
X.head(4)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=32,stratify=y)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
svc = SVC()
svc.fit(X_train_scaled, y_train)
y_pred = svc.predict(X_test_scaled)

print("Confusion Matrix :\n", confusion_matrix(y_test, y_pred))
print("Accuracy Score :", accuracy_score(y_test, y_pred))
print("Precision Score :", precision_score(y_test, y_pred))
print("Recall Score :", recall_score(y_test, y_pred))
print("F1 Score :", f1_score(y_test, y_pred))
```

```python
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import GridSearchCV

param_grid = {
    "C" : [0.1,0.001,0.001,1],
    "gamma" : [0.0001,0.1,0.01,"scale"],
    "kernel" : ["poly","linear"]
}

grid = GridSearchCV(SVC(), param_grid, cv= 4, refit= True)
grid.fit(X_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_prid= grid.predict(X_test_scaled)

print("Confusion Matrix :\n", confusion_matrix(y_test, grid_prid))
print("Accuracy Score :", accuracy_score(y_test, grid_prid))
print("Precision Score :", precision_score(y_test, grid_prid))
print("Recall Score :", recall_score(y_test, grid_prid))
print("F1 Score :", f1_score(y_test, grid_prid))
```

### Adding Categorical Features from here


```python
numeric_features = [
    "Age",
    "AnnualIncome",
    "CreditScore",
    "LoanAmount",
    "LoanDuration",
    "MonthlyDebtPayments",
    "DebtToIncomeRatio",
    "TotalDebtToIncomeRatio",
    "NetWorth",
    "RiskScore"
]
categorical_features = [
    "EmploymentStatus",
    "EducationLevel",
    "MaritalStatus",
    "HomeOwnershipStatus",
    "LoanPurpose"
]
```

```python
X_full = df[ numeric_features + categorical_features]
y_full = df["LoanApproved"]
```

```python
df_full = pd.concat([X_full,y_full],axis=1)
df_full.head(3)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_full, y_full, test_size=0.2, random_state=32,stratify=y_full)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(
    transformers=[
        ("onehot", OneHotEncoder(drop= "first"),categorical_features)],
        remainder= "passthrough")

X_train_encoded = ct.fit_transform(X_train)
X_test_encoded = ct.transform(X_test)
```

```python
feature_names = ct.get_feature_names_out()
X_train_encoded_df = pd.DataFrame(X_train_encoded, columns=feature_names)
display(X_train_encoded_df.head())
```

```python
import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
print(X_train_encoded)
```

```python
scaler1 = StandardScaler()
X_train_full_scaled = scaler1.fit_transform(X_train_encoded)
X_test_full_scaled = scaler1.transform(X_test_encoded)
```

```python
svc1 = SVC()
svc1.fit(X_train_full_scaled,y_train)
y_pred_full = svc1.predict(X_test_full_scaled)

print("Confusion Matrix :\n", confusion_matrix(y_test, y_pred_full))
print("Accuracy Score :", accuracy_score(y_test, y_pred_full))
print("Precision Score :", precision_score(y_test, y_pred_full))
print("Recall Score :", recall_score(y_test, y_pred_full))
```

```python

param_grid1 = {
    "C" : [0.1,0.001,1,10],
    "gamma" : [0.0001,0.1,0.01,"scale"],
    "kernel" : ["poly","linear","rbf"]
}
grid_full = GridSearchCV(SVC(), param_grid1, cv= 4, refit= True)
grid_full.fit(X_train_full_scaled,y_train)
grid_full_pred= grid_full.predict(X_test_full_scaled)
print("Confusion Matrix :\n", confusion_matrix(y_test, grid_full_pred))
print("Accuracy Score :", accuracy_score(y_test, grid_full_pred))
print("Precision Score :", precision_score(y_test, grid_full_pred))
print("Recall Score :", recall_score(y_test, grid_full_pred))
print("F1 Score :", f1_score(y_test, grid_full_pred))
```
