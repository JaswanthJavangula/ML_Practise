```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = load_breast_cancer()
df= pd.DataFrame(data.data, columns=data.feature_names)
df.columns
```

```python
X= data.data
y= data.target
print(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,stratify=y, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled =scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.svm import SVC
model = SVC(kernel = "rbf"
)

model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
model = SVC(kernel = "linear"
)

model.fit(X_train_scaled, y_train)
y_pred1 = model.predict(X_test_scaled)
print(confusion_matrix(y_test, y_pred1))
print(accuracy_score(y_test, y_pred1))
print(classification_report(y_test, y_pred1))
```

```python
model = SVC(kernel = "sigmoid"
)
model.fit(X_train_scaled, y_train)
y_pred2 = model.predict(X_test_scaled)
print(confusion_matrix(y_test, y_pred2))
print(accuracy_score(y_test, y_pred2))
print(classification_report(y_test, y_pred2))
```

```python
model = SVC(kernel = "poly"
)

model.fit(X_train_scaled, y_train)
y_pred3 = model.predict(X_test_scaled)
print(confusion_matrix(y_test, y_pred3))
print(accuracy_score(y_test, y_pred3))
print(classification_report(y_test, y_pred3))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    "C" : [0.1, 1, 10, 100],
    "gamma" : [1, 0.1, 0.01, 0.001],
    "kernel" : ["poly"]
    }
grid = GridSearchCV(SVC(), param_grid,  verbose= 3,refit= True)

grid.fit(X_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
y_pred4 = grid.predict(X_test_scaled)
print(classification_report(y_test, y_pred4))
print(confusion_matrix(y_test, y_pred4))
print(accuracy_score(y_test, y_pred4))
```

### Visualizing Linear Kernel SVM Decision Boundary (Training Set)

Let's visualize the decision boundary of the Linear Kernel SVM on the training data, focusing on the first two features and setting others to their mean. The classification regions will be shown, along with the individual training points.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from matplotlib.colors import ListedColormap
from sklearn.datasets import load_breast_cancer # Ensure data is loaded for feature_names

sns.set_style("whitegrid") # Set seaborn style for better aesthetics

data = load_breast_cancer() # Re-load data to access feature_names

# Prepare data for 2D visualization (unscaled first two features of training data)
X_set_plot_train = X_train[:, :2]
y_set_plot_train = y_train

# Make predictions on the training data
y_pred_train_linear = model.predict(X_train_scaled)

# Create a DataFrame for plotting to easily use seaborn's hue and style
plot_df_train = pd.DataFrame(X_set_plot_train, columns=[data.feature_names[0], data.feature_names[1]])
plot_df_train['True Label'] = y_set_plot_train
plot_df_train['Predicted Label'] = y_pred_train_linear
plot_df_train['Classification'] = np.where(plot_df_train['True Label'] == plot_df_train['Predicted Label'], 'Correct', 'Incorrect')

# Define ranges for the meshgrid based on the first two features of the training set
X1_min, X1_max = X_set_plot_train[:, 0].min() - 1, X_set_plot_train[:, 0].max() + 1
X2_min, X2_max = X_set_plot_train[:, 1].min() - 1, X_set_plot_train[:, 1].max() + 1

# Create a meshgrid
X1, X2 = np.meshgrid(np.arange(start=X1_min, stop=X1_max, step=0.05),
                     np.arange(start=X2_min, stop=X2_max, step=0.05))

# Create a dummy array for the remaining 28 features, set to their mean from X_train
remaining_features_mean = X_train[:, 2:].mean(axis=0)

# Construct the full feature array for prediction on the meshgrid points
meshgrid_points_2d = np.array([X1.ravel(), X2.ravel()]).T
remaining_features_meshgrid = np.tile(remaining_features_mean, (meshgrid_points_2d.shape[0], 1))
full_meshgrid_points = np.hstack((meshgrid_points_2d, remaining_features_meshgrid))

# Scale the full meshgrid points using the already fitted scaler
scaled_full_meshgrid_points = scaler.transform(full_meshgrid_points)

# Predict on the scaled full meshgrid points using the linear model (from cell tFyHudFKeATZ)
Z_linear_train = model.predict(scaled_full_meshgrid_points).reshape(X1.shape)

# Plotting
plt.figure(figsize=(10, 7))
# Red for Class 0, Green for Class 1 for contourf (as per user's example)
plt.contourf(X1, X2, Z_linear_train, alpha=0.75, cmap=ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

# Plot the training data points with hue for true label and style for classification accuracy
sns.scatterplot(x=data.feature_names[0],
                y=data.feature_names[1],
                hue='True Label',
                style='Classification',
                markers={'Correct': 'o', 'Incorrect': 'X'}, # Use different markers for correct/incorrect
                palette={0: 'red', 1: 'green'}, # Consistent with contourf cmap
                s=100,
                alpha=0.8,
                edgecolor='k',
                data=plot_df_train)

plt.title('Linear SVM Decision Boundary (Training Set) - Correct/Incorrect Classification')
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
