```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification

X,y = make_classification(
    n_samples = 2500,n_features=5,n_redundant=1,n_classes=2,random_state=123
    )
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y, test_size =0.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNC=  KNeighborsClassifier(n_neighbors=4,algorithm="auto")
KNC.fit(X_train,y_train)
y_pred = KNC.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
accuracy_score(y_test,y_pred)
```

grid search cv

```python
from sklearn.model_selection import GridSearchCV

param_grid ={
    "n_neighbors" : [1,2,3,4,6,7,8,9]
}
grid = GridSearchCV(KNC, param_grid,cv=5,verbose=1,n_jobs=-1)
grid.fit(X_train,y_train)
y_pred_grid = grid.predict(X_test)
print(grid.best_params_)

print(classification_report(y_test,y_pred_grid))
print(confusion_matrix(y_test,y_pred_grid))
accuracy_score(y_test,y_pred_grid)
```
