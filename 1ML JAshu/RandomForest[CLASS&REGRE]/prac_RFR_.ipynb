```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "train.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "jaswanthjavangula/housing-regression",
  file_path,
)

print("First 5 records:", df.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.DataFrame(df)
df.drop("Id",axis=1,inplace=True)
df.head(3)
```

```python
y = df["SalePrice"]
X = df.drop("SalePrice",axis=1)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.33, random_state=22)

num_features = [i for i in X.columns if X[i].dtypes != "O"]
cat_features = [i for i in X.columns if i not in num_features]
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, root_mean_squared_error, mean_absolute_error

numeric_TRANSFORMER = Pipeline(
    steps = [
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ]
        )

cat_transformer = Pipeline(
    steps = [
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("oneHot", OneHotEncoder(handle_unknown = "ignore"))
    ]
)

preprocess = ColumnTransformer(
    transformers=[
    ("cat_trans", cat_transformer, cat_features),
    ("num_tran", numeric_TRANSFORMER, num_features)
])

rf = RandomForestRegressor(max_depth=6, n_estimators=60,random_state=42)
```

```python
rf_pipeline = Pipeline(
    steps= [
        ("preprocessing", preprocess),
        ("model", rf)
    ]
)

rf_pipeline.fit(X_train, y_train)

y_train_pred = rf_pipeline.predict(X_train)
y_val_pred = rf_pipeline.predict(X_val)

lr_pipeline = Pipeline(
    steps= [
        ("preprocessing", preprocess),
        ("model", LinearRegression())
    ]
)
```

```python
# train_eval(y_train, y_train_pred)

# val_eval(y_val, y_val_pred)
```

```python
def eval_met (actual, pred):
  rmse = np.sqrt(mean_squared_error(actual, pred))
  mae = mean_absolute_error(actual, pred)
  r2 = r2_score(actual, pred)
  return rmse, mae, r2
```

```python
models ={
    "Random Forest Pipeline" :rf_pipeline,
    "Linear Regression" :lr_pipeline
}

for i in range(len(list(models))):
  model = list(models.values())[i]
  model.fit(X_train,y_train)

  y_train_pred1 = model.predict(X_train)
  y_val_pred1 = model.predict(X_val)

  print(list(models.keys())[i])

  train_rmse, train_mae , train_r2 = eval_met(y_train,y_train_pred1)
  val_rmse, val_mae,val_r2 = eval_met(y_val,y_val_pred1)

  print("Train RMSE:",train_rmse)
  print("Train MAE:",train_mae)
  print("Train R2:",train_r2)
  print("-"*50)
  print("Val RMSE:",val_rmse)
  print("Val MAE:",val_mae)
  print("Val R2:",val_r2)
  print("*"*50)
```
