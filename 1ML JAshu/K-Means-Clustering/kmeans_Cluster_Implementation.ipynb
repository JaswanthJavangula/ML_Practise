```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
```

```python
X,y = make_blobs(n_samples=1000, centers=3 , n_features=2)
```

```python
X,y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.preprocessing import StandardScaler
src = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train_scaled= src.fit_transform(X_train)
X_test_scaled = src.transform(X_test)
```

```python
from sklearn.cluster import KMeans
```

```python
# elbow method
wcss= []
for i in range(1,11):
  kmeans= KMeans(n_clusters=i, init= "k-means++")
  kmeans.fit(X_train_scaled)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
# plot elbow curve
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
k_mean = KMeans(n_clusters=3, init= "k-means++")
k_mean.fit_predict(X_train_scaled)
```

```python
y_pred = k_mean.predict(X_test_scaled)
y_pred
```

```python
plt.scatter(X_test_scaled[:,0],X_test_scaled[:,1],c=y_pred)
```

```python
# second method
# # silhoutte scoring
from sklearn.metrics import silhouette_score
```

```python
Sil_coeff = []
for k in range(2,11):
  kmeans = KMeans(n_clusters=k , init = "k-means++")
  kmeans.fit(X_train_scaled)
  score = silhouette_score(X_train_scaled, kmeans.labels_)
  Sil_coeff.append(score)
```

```python
Sil_coeff
```

```python
# plotting silhoutte
plt.plot(range(2,11),Sil_coeff)
plt.xticks(range(2,11))
plt.xlabel("Number of clusters")
plt.ylabel("Silhoutte Coefficient")
plt.show()
```
